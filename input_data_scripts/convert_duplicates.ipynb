```python
import pandas as pd
import json
```

```python
stops = {}

with (open("stations.json", "r") as f):
    stops = json.load(f)
```

```python
duplicates = []

#Linear Double Search (O(n^2) lol)

for stopA, stopdataA in stops.items():
    nameA = stopdataA["name"]

    for stopB, stopdataB in stops.items():
        nameB = stopdataB["name"]

        #If they hae the same base name, minus bus
        if (stopA != stopB):
            if (nameA.replace(" Bus", "") == nameB.replace(" Bus", "") and nameA != nameB and "GO" in nameA):
                #print(nameA, nameB)
                #print((stopA, stopB))
                if ("Bus" in nameA):
                    duplicates.append(stopA);
                    duplicates.append(stopB);
                else:
                    duplicates.append(stopB);
                    duplicates.append(stopA);

duplicates
```

```python
with (open("duplicateKeys.json", "w") as f):

    duplicates2 = []

    i = 0;
    while i < len(duplicates):
        duplicates2.append([duplicates[i], duplicates[i+1]])
        i += 2
    
    finalwrite = json.dumps(duplicates2, indent = 4)
    f.write(finalwrite)
```

```python
endjson = {}

duplicates3 = duplicates[::]

for stopid, stopdata in stops.items():

    isADup = False

    #Decide if It is a duplicate (but only if it is the bus component)
    if stopid in duplicates:

        #Check
        ind = duplicates.index(stopid)
        
        #Check If I Should Remove
        if (ind % 2 == 0):
            isADup = True
            duplicates.pop(ind)
            print(stopid + " removed." + str(ind))

    if (not isADup):
        endjson[stopid] = stopdata

with (open("stations_no_dups.json", "w") as f):
    finalwrite = json.dumps(endjson, indent = 4)
    f.write(finalwrite)
```

```python
#Modify the Route Edges
with open("route_edges.json", "r") as f:
    lines = f.readlines();

newlines = []

for line in lines:

    for duplicate in duplicates3:

        #Check
        ind = duplicates3.index(duplicate)

        if ind % 2 == 0:
            stopAinRoute = "\"{}-".format(duplicate) in line
            stopBinRoute = "-{}\"".format(duplicate) in line
            stopAinRouteNode = "\"node1\": \"{}".format(duplicate) in line
            stopBinRouteNode = "\"node2\": \"{}".format(duplicate) in line

            if stopAinRoute or stopBinRoute or stopAinRouteNode or stopBinRouteNode:
                new = duplicates3[ind+1]

                print(duplicate + " found")
                print("replaceing with " + new)
                line = line.replace(duplicate, new)
                break

    newlines.append(line)
        
with open("route_edges_combined.json", "w") as f:
    f.writelines(newlines)
```
